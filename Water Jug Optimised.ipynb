```python
tree = {}

jug_1_max_cap = 3
jug_2_max_cap = 5

jug_1_final_cap = 0
jug_2_final_cap = 2

s = 0
```

```python
class State:
    def __init__(self, x, y, Id, parent, children):
        self.x = x
        self.y = y
        self.id = Id
        self.parent = parent
        self.children = children
    def fillX(self):
        self.x = jug_1_max_cap
    def fillY(self):
        self.y = jug_2_max_cap
    def fillXFromY(self):
        if(self.y>=jug_1_max_cap-self.x):
            self.y = self.y - (jug_1_max_cap-self.x)
            self.x = jug_1_max_cap
        else:
            self.x = self.x + self.y
            self.y = 0
    def fillYFromX(self):
        if(self.x>=jug_2_max_cap-self.y):
                self.x = self.x - (jug_2_max_cap-self.y)
                self.y = jug_2_max_cap
        else:
            self.y = self.y + self.x
            self.x = 0
    def EmptyX(self):
        self.x = 0
    def EmptyY(self):
        self.y = 0
    def hasReachedFinal(self):
        if(self.x==jug_1_final_cap and self.y==jug_2_final_cap):
                return True
        return False
```

```python
visited_states=[[0,0]]
```

```python
root = State(0,0,s,None,[])

tree[s] = root
parent = 0
s += 1

while(True):
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[])
    if(newNode.y != jug_2_max_cap):
        newNode.fillY()
        if([newNode.x,newNode.y] not in visited_states):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            visited_states.append([newNode.x,newNode.y])
            s += 1
            if(newNode.hasReachedFinal()):
                break
            
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[])
    if(newNode.x != jug_1_max_cap):
        newNode.fillX()
        if([newNode.x,newNode.y] not in visited_states):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            visited_states.append([newNode.x,newNode.y])
            s += 1
            if(newNode.hasReachedFinal()):
                break
            
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[])
    if(newNode.y != 0):
        newNode.EmptyY()
        if([newNode.x,newNode.y] not in visited_states):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            visited_states.append([newNode.x,newNode.y])
            s += 1
            if(newNode.hasReachedFinal()):
                break
            
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[])
    if(newNode.x != 0):
        newNode.EmptyX()
        if([newNode.x,newNode.y] not in visited_states):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            visited_states.append([newNode.x,newNode.y])
            s += 1
            if(newNode.hasReachedFinal()):
                break
            
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[])
    if(newNode.x != 0 and newNode.y!=jug_2_max_cap):
        newNode.fillYFromX()
        if([newNode.x,newNode.y] not in visited_states):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            visited_states.append([newNode.x,newNode.y])
            s += 1
            if(newNode.hasReachedFinal()):
                break
            
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[])
    if(newNode.y != 0 and newNode.x!=jug_1_max_cap):
        newNode.fillXFromY()
        if([newNode.x,newNode.y] not in visited_states):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            visited_states.append([newNode.x,newNode.y])
            s += 1
            if(newNode.hasReachedFinal()):
                break
            
    parent +=1
    
```

```python
for i in tree:
    print(str(i) + " - " + str(tree[i].parent) + " : " + str(tree[i].x) + "," + str(tree[i].y))
```
